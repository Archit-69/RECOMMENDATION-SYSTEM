## Recommendation System
This notebook builds a movie recommendation system using Collaborative Filtering (Matrix Factorization).

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import surprise
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split as sp_split
```

## Load Sample Dataset (MovieLens)

```python
data = surprise.Dataset.load_builtin('ml-100k')
trainset, testset = sp_split(data, test_size=0.2)
trainset, testset
```

## Train Matrix Factorization Model (SVD)

```python
model = SVD()
model.fit(trainset)
predictions = model.test(testset)
predictions[:5]
```

## Evaluate Model (RMSE)

```python
rmse = surprise.accuracy.rmse(predictions)
```

## Make Recommendation for a User

```python
user_id = '10'
all_items = trainset.all_items()
item_ids = [trainset.to_raw_iid(i) for i in all_items]
        # collect items the user has already rated (item/raw id is at index 1 in testset tuples)
        user_rated = [j[1] for j in testset if j[0] == user_id]
recs = []
for iid in item_ids:
            # skip items already rated by the user
            if iid in user_rated:
                continue
            est = model.predict(user_id, iid).est
            recs.append((iid, est))
top10 = sorted(recs, key=lambda x:x[1], reverse=True)[:10]
top10
```
